```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import pickle
```

```python
with open('texts.pickle', 'rb') as handle:
    texts = pickle.load(handle)
with open('labels.pickle', 'rb') as handle:
    labels = pickle.load(handle)
```

```python
# distribution of tweet lengths
lengths = np.array([len(row.split(" ")) for row in texts])
summary = "mean: " + str(int(np.mean(lengths))) + " , min/max: " + str(np.min(lengths))+"/"+str(np.max(lengths))
plt.figure(1, figsize=(10,4))
plt.hist(lengths, bins='auto')
plt.title("Distribution of word counts")
plt.xlim(0, 300)
plt.xlabel("Text Length: " + summary); plt.ylabel("Examples")
plt.axvline(np.mean(lengths), ls="--", color="b")
plt.axvline(np.percentile(lengths, 95), ls="--", color="y")
plt.show()
# distribution of label counts
labels = np.array(labels)
label_counts = [np.count_nonzero(labels[:,i]) for i in range(7)]
plt.figure(2, figsize=(10,4))
classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate", "clean"]
plt.bar(classes[:-1],label_counts[:-1])
plt.title("Distribution of labels (excluding clean)")
plt.xlabel("Labels"); plt.ylabel("Examples")
plt.show()
```
